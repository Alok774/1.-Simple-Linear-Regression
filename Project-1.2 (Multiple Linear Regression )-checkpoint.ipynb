**Step-3.4: Tain-test Split**

```python
X=df.drop(columns='sales')
y=df['sales']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=50)
```

# step-4: Modeling

**Multiple Linear Regression-** 
* Linear regression include multiple input features.This is done by least squares method
y=B0+B1x1+..+Bnxn
* Each x represents a different feature,and each feature has its own cofficient.
**In this case:**

y=B0 + B1*TV + B2*Radio + B3*Newspaper

```python
from sklearn.linear_model import LinearRegression #import
model=LinearRegression()                          #instantiate
model.fit(X_train,y_train)                        #fit
```

```python
model.intercept_
```

```python
model.coef_
```

# Prediction

```python
train_prediction=model.predict(X_train)
test_prediction=model.predict(X_test)
```

# Step-5:Evaluation Matrics

```python
from sklearn.metrics import mean_squared_error
test_RMSE = np.sqrt(mean_squared_error(y_test,test_prediction))
train_RMSE = np.sqrt(mean_squared_error(y_train,train_prediction))
print(train_RMSE,test_RMSE)
```

```python
model.score(X_train,y_train) #Train R2
```

```python
model.score(X_test,y_test)  # Test R-Square
```

* alternative hypothesis:

**Variance Influence Factor (VIF)**

* VIF measures the radio between the variance for a given regression cofficient with only that variable in the model versus the variance for a given regression cofficient with all variables in the model.

```python
rsq_TV=smf.ols('TV~radio+newspaper',data=df).fit()
rsq_TV.summary()
```

```python
# calculateing VIF's values of independent variables
rsq_TV=smf.ols('TV~radio+newspaper',data=df).fit().rsquared
vif_TV=1/(1-rsq_TV)

rsq_radio=smf.ols('radio~TV+newspaper',data=df).fit().rsquared
vif_radio=1/(1-rsq_radio)

rsq_newspaper=smf.ols('newspaper~TV+radio',data=df).fit().rsquared
vif_newspaper=1/(1-rsq_newspaper)
```

```python
# Soorting vif values in a data frame
d1={'Variables':['TV','radio','newspaper'],'VIF':[vif_TV,vif_radio,vif_newspaper]}
vif_frame=pd.DataFrame(d1)
vif_frame
```

* If VIF > 4, for any independent variable,then drop that particular variable
* If VIF of model > 4 (threshold value=4)..then there exists a problem and reinvestigate the problem
